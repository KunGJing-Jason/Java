```python
import pandas as pd
import numpy as np
import math

# 读取数据
data = pd.read_excel('附件一：已结束项目任务数据.xls')
trd = pd.read_excel('附件二：会员信息数据.xlsx')

# 获取任务A0001和A0002的经纬度
x1 = data.loc[data['任务号码'] == 'A0001', '任务gps纬度'].values[0]
y1 = data.loc[data['任务号码'] == 'A0001', '任务gps经度'].values[0]
x2 = data.loc[data['任务号码'] == 'A0002', '任务gps纬度'].values[0]
y2 = data.loc[data['任务号码'] == 'A0002', '任务gps经度'].values[0]

# 获取会员的经纬度
b = trd['会员位置(GPS)'].str.extract('(\d+\.\d+)\s(\d+\.\d+)', expand=True).astype(float)
z1 = trd['会员编号'].values
x4 = b.iloc[:, 0].values
y4 = b.iloc[:, 1].values

# 计算会员与任务A0001和A0002的距离
D1 = 111.199 * np.sqrt((x1 - x4) ** 2 + (y1 - y4) ** 2 * np.cos((x1 + x4) * math.pi / 180) ** 2)
D2 = 111.199 * np.sqrt((x2 - x4) ** 2 + (y2 - y4) ** 2 * np.cos((x2 + x4) * math.pi / 180) ** 2)

# 将会员与任务A0001和A0002的距离转化为Series类型，以会员编号为索引
s_A0001 = pd.Series(D1, index=z1)
s_A0002 = pd.Series(D2, index=z1)

# 统计任务A0001在5公里范围内的会员个数
A0001_Bnum = len(D1[D1 <= 5])

# 统计任务A0002在5公里范围内的会员信誉值总和
D3 = s_A0002[s_A0002 <= 5]
A0002_Bavg = trd.loc[trd['会员编号'].isin(D3.index), '信誉值'].sum()

# 输出结果和问题描述
q_1 = '1) 计算A0001、A0002任务位置到所有会员位置的距离，结果采用两个序列来保存，分别记为S_A0001、S_A0002，其中index为会员编号，值为其对应的距离。'
q_2 = '2) 计算A0001任务在5公里范围内的会员个数，记为A0001_Bnum。'
q_3 = '3) 计算A0002任务在5公里范围内所有会员信誉值总和，记为A0002_Bavg。'

print(q_1, '\n', s_A0001, s_A0002)
print(q_2, '\n', A0001_Bnum)
print(q_3, '\n', A0002_Bavg)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# 读取数据
data = pd.read_excel('银行贷款审批数据.xlsx')

# 对前6列数据进行缺失值填充和标准化
d1 = data.iloc[:, 0:6]
imp1 = SimpleImputer(strategy='mean')
d1 = imp1.fit_transform(d1)
scaler1 = StandardScaler()
d1 = scaler1.fit_transform(d1)

# 对后9列数据进行缺失值填充
d2 = data.iloc[:, 6:15]
imp2 = SimpleImputer(strategy='median')
d2 = imp2.fit_transform(d2)

# 将前6列和后9列数据合并
D = np.concatenate((d1, d2), axis=1)

# 划分训练集和测试集
x = D[:600, :]
y = data.iloc[:600, 15]
x1 = D[600:, :]
y1 = data.iloc[600:, 15]

# 创建支持向量机模型并训练
clf_svm = svm.SVC(kernel='rbf')
clf_svm.fit(x, y)
rv_svm = clf_svm.score(x, y)
R1 = clf_svm.predict(x1)
Z1 = R1 - y1
Rs1 = len(Z1[Z1==0]) / len(Z1)

# 创建逻辑回归模型并训练
clf_lr = LogisticRegression()
clf_lr.fit(x, y)
rv_lr = clf_lr.score(x, y)
R2 = clf_lr.predict(x1)
Z2 = R2 - y1
Rs2 = len(Z2[Z2==0]) / len(Z2)

# 创建神经网络模型并训练
clf_nn = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)
clf_nn.fit(x, y)
rv_nn = clf_nn.score(x, y)
R3 = clf_nn.predict(x1)
Z3 = R3 - y1
Rs3 = len(Z3[Z3==0]) / len(Z3)

# 打印模型预测结果和精度
print('支持向量机模型预测结果为：', R1)
print('逻辑回归模型预测结果为：', R2)
print('神经网络模型预测结果为：', R3)
print('支持向量机模型预测精度为：', Rs1)
print('逻辑回归模型预测精度为：', Rs2)
print('神经网络模型预测精度为：', Rs3)

# 绘制模型预测精度柱状图
hz = ['支持向量机', '逻辑回归', '神经网络']
zz = [Rs1, Rs2, Rs3]
plt.figure(1)
font = {'family': 'SimHei'}
plt.rc('font', **font)
plt.bar(hz, zz)
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans 

# 读取数据
data = pd.read_excel('高教数据.xlsx')

# 提取特征数据并进行规范化处理
X = data.iloc[:, 1:]
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 进行主成分分析
pca = PCA(n_components=0.90) # 累计贡献率达到90%以上
Y = pca.fit_transform(X) # 主成分

# 进行K-均值聚类
model = KMeans(n_clusters=4, random_state=0, max_iter=500)
model.fit(Y)
c = model.labels_ # K-均值聚类结果

# 整理K-均值聚类结果，并打印出各类别的地区名
Fs = pd.Series(c, index=data['地区'])
Fs = Fs.sort_values(ascending=True)
for i in range(4):
    print(f'第{i}类为：', list(Fs[Fs==i].index))
```

```python
import pandas as pd
import numpy as np
from apriori import find_rule # 导入自行编写的apriori函数
import warnings
warnings.filterwarnings('ignore')

# 商品列表
tiem = ['西红柿', '排骨', '鸡蛋', '毛巾', '水果刀', '茄子', '香蕉', '袜子', '肥皂', '酸奶', '土豆', '鞋子']

# 读取数据
data = pd.read_table('第四题.txt', sep='、', header=None, encoding='ansi', engine='python')

# 构建字典，记录每个商品在每个购物篮中是否出现
D = {}
for t in range(len(tiem)):
    z = np.zeros((len(data)))
    li = []
    for k in range(len(data.iloc[0, :])):
        s = data.iloc[:, k] == tiem[t]
        li.extend(list(s[s.values == True].index))
    z[li] = 1
    D.setdefault(tiem[t], z)

# 将字典转换为DataFrame格式
Data = pd.DataFrame(D)

# 调用apriori函数进行关联规则挖掘并将结果保存到excel文件中
output_file = 'apriori_rules.xls' # 结果文件
support = 0.2 # 最小支持度
confidence = 0.4 # 最小置信度
ms = '---' # 连接符，默认'--'
rules = find_rule(Data, support, confidence, ms)
rules.to_excel(output_file)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC

# 读取数据
td = pd.read_excel('沪深300指数2014年交易数据.xlsx')

# 获取指标
A1 = td['Idxtrd05'].values / td['Idxtrd05'].rolling(10).mean()
A2 = td['Idxtrd06'].values / td['Idxtrd06'].rolling(10).mean()
A4 = td['Idxtrd03'].values / td['Idxtrd05'].rolling(10).mean()
A5 = td['Idxtrd04'].values / td['Idxtrd05'].rolling(10).mean()
A6 = td['Idxtrd03'].values - td['Idxtrd04'].values
A7 = td['Idxtrd05'].values - td['Idxtrd02'].values
Y = td['Idxtrd05'].values[1:] - td['Idxtrd05'].values[:-1]
A3 = np.zeros(len(td))
for i in range(len(td)):
    if i > 0:
        A3[i] = (td.loc[i, 'Idxtrd05'] - td.loc[i-1, 'Idxtrd05']) / td.loc[i-1, 'Idxtrd05']
X = {'A1': A1, 'A2': A2, 'A3': A3, 'A4': A4, 'A5': A5, 'A6': A6, 'A7': A7}
X = pd.DataFrame(X)
X = X.iloc[9:,]

# 对指标A1~A7作标准化处理
scaler = StandardScaler()
X = scaler.fit_transform(X)

Y = Y[8:]
Y[Y > 0] = 1
Y[Y <= 0] = -1
Y = Y.reshape(len(Y), 1)

# 划分训练集和测试集
x = X[:len(X)-30, :]
y = Y[:len(Y)-30, :]
x1 = X[len(X)-30:, :]
y1 = Y[len(Y)-30:, :]

# 支持向量机
clf = SVC(kernel='rbf')
clf.fit(x, y.ravel())
train_score = clf.score(x, y)  # 训练集准确率
test_score = clf.score(x1, y1)  # 测试集准确率
predict = clf.predict(x1)
accuracy = np.sum(predict == y1) / len(y1)  # 预测准确率

print('训练集准确率为：', train_score)
print('测试集准确率为：', test_score)
print('预测准确率为：', accuracy)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.svm import SVC
import warnings
warnings.filterwarnings('ignore')

# 读取数据
A = pd.read_excel('hepatitis.xlsx')

# 缺失值"?"转换为nan
A.replace('?', np.nan, inplace=True)

# B1均值策略填充及数据标准化
B1 = A.iloc[:, :6].copy()
imp = SimpleImputer(strategy='mean')
B1 = imp.fit_transform(B1)
min_max_scaler = MinMaxScaler()
B1 = min_max_scaler.fit_transform(B1)

# B2最频繁值填充
B2 = A.iloc[:, 6:-1].copy()
imp = SimpleImputer(strategy='most_frequent')
B2 = imp.fit_transform(B2)

# 合并B1和B2,并做主成分分析，提取主成分
B = np.hstack((B1, B2))
pca = PCA(n_components=0.95)
X = pca.fit_transform(B)

# 决策树预测
x = X[:130, :]
x1 = X[130:, :]
y = A.iloc[:, -1][:130].values
y1 = A.iloc[:, -1][130:].values

clf = SVC()  
clf.fit(x, y)
y0 = clf.predict(x1)
accuracy = np.sum(y0 == y1) / len(y1)

print("决策树模型预测准确率为：", accuracy)
```

```python
import pandas as pd
import numpy as np
import math

# 导入自定义函数模块
import fun

# 读取数据
A = pd.read_excel('附件一：已结束项目任务数据.xls')
B = pd.read_excel('附件二：会员信息数据2.xlsx')

# 初始化结果矩阵
Z = np.zeros((len(A), 13))

# 计算每个任务的相关指标
for t in range(len(A)):
    # 获取任务的经纬度
    A_W0 = A.iloc[t, 1]
    A_J0 = A.iloc[t, 2]

    # 计算任务与其他任务的距离
    Z1 = np.zeros(len(A))
    for i in range(len(A)):
        A_Wi = A.iloc[i, 1]
        A_Ji = A.iloc[i, 2]
        d1 = 111.19 * math.sqrt((A_W0 - A_Wi) ** 2 + (A_J0 - A_Ji) ** 2 * math.cos((A_W0 + A_Wi) * math.pi / 180) ** 2)
        Z1[i] = d1

    # 计算任务与会员的距离
    Z2 = np.zeros(len(B))
    for k in range(len(B)):
        B_Wk = B.iloc[k, 1]
        B_Jk = B.iloc[k, 2]
        d2 = 111.19 * math.sqrt((A_W0 - B_Wk) ** 2 + (A_J0 - B_Jk) ** 2 * math.cos((A_W0 + B_Wk) * math.pi / 180) ** 2)
        Z2[k] = d2

    # 计算任务的相关指标
    Z[t, 0] = t
    Z[t, 1] = len(Z1[Z1 <= 5])
    Z[t, 2] = A.iloc[Z1 <= 5, 3].mean()
    Z[t, 3] = B.iloc[Z2 <= 5, 3].sum()
    Z[t, 4] = B.iloc[Z2 <= 5, 5].mean()
    Z[t, 5] = len(Z2[Z2 <= 5])
    Z[t, 6] = B.iloc[fun.find_I(6, 30, 6, 30, Z2, B), 3].sum()
    Z[t, 7] = B.iloc[fun.find_I(6, 33, 6, 45, Z2, B), 3].sum()
    Z[t, 8] = B.iloc[fun.find_I(6, 48, 7, 3, Z2, B), 3].sum()
    Z[t, 9] = B.iloc[fun.find_I(7, 6, 7, 21, Z2, B), 3].sum()
    Z[t, 10] = B.iloc[fun.find_I(7, 24, 7, 39, Z2, B), 3].sum()
    Z[t, 11] = B.iloc[fun.find_I(7, 42, 7, 57, Z2, B), 3].sum()
    Z[t, 12] = B.iloc[fun.find_I(8, 0, 8, 0, Z2, B), 3].sum()

# 保存结果矩阵
np.save('Z', Z)

# 打印结果矩阵
print(Z)
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.neural_network import MLPRegressor
from sklearn import svm
import warnings
warnings.filterwarnings('ignore')

# 导入数据
Z = np.load('Z.npy')
A = pd.read_excel('附件一：已结束项目任务数据.xls')

# 提取特征和目标变量
Data = pd.DataFrame(Z[:, 1:])
A4 = A.iloc[:, 4].values
y = A.iloc[:, 3].values
y = y.reshape(len(y), 1)
y0 = y[A4 == 0]
y = y[A4 == 1]

# 数据预处理
X = Data.fillna(0).to_numpy()
scaler = StandardScaler()
X = scaler.fit_transform(X)
pca = PCA(n_components=0.9)
Y = pca.fit_transform(X)
tzxl = pca.components_
tz = pca.explained_variance_
gxl = pca.explained_variance_ratio_

# 多层感知器回归
x = Y[A4 == 1, :]
x1 = Y[A4 == 0, :]
clf = MLPRegressor(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=35, random_state=1)
clf.fit(x, y)
rv = clf.score(x, y)
y1 = clf.predict(x1)

# 支持向量机分类
xx = np.concatenate((X, A.iloc[:, [3]]), axis=1)
yy = A4.reshape(len(A4), 1)
clf = svm.SVC(kernel='rbf')
clf.fit(xx, yy)
rv = clf.score(xx, yy)
xx1 = np.hstack((Z[A4 == 0, 1:], y1.reshape(len(y1), 1)))
R = clf.predict(xx1)

# 计算结果
R1 = len(R[R==1])      #任务完成增加量
R2 = sum(y1) - sum(y0) #成本增加额

# 输出结果
print('任务完成增加量: \n', R1)
print('成本增加额: \n', R2)
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier

# 读取Excel文件中的数据
A = pd.read_excel('2018年财务指标数据.xlsx')
df = A.dropna()

# 将Stkcd列设置为数据框的索引
df = df.set_index('Stkcd')

# 区分ST公司和非ST公司
# ST公司：选出前四个指标中任意一个小于0的公司，添加到B1，并在B1添加列标为1
# 非ST公司：选出前四个指标都大于0的公司，添加到B，并在B添加列标为0
B1 = df[(df.iloc[:, :4] < 0).any(axis=1)].iloc[:, 4:]
B1['y'] = 1
B = df[(df.iloc[:, :4] >= 0).all(axis=1)].iloc[:, 4:]
B = B.sample(n=len(B1), replace=True, random_state=10, axis=0)
B['y'] = 0

# 合并B1和B，并提取出这些样本公司的代码
f0 = pd.concat([B, B1], axis=0)
B2 = pd.DataFrame(f0.index)

# 0-1规范化
min_max_scaler = MinMaxScaler()
XZ = min_max_scaler.fit_transform(f0.iloc[:, :14])
f1 = pd.DataFrame(XZ)

# 合并B2，f1和f0的y列
f = pd.concat([B2, f1, pd.DataFrame(np.array(f0.iloc[:, 14]))], axis=1)

# 提取95%的主成分
pca = PCA(n_components=0.95)
pca.fit(f.iloc[:, 1:15])# 对数据进行拟合训练
Y = pca.transform(f.iloc[:, 1:15])# 返回提取的主成分

# 神经网络模型
x = Y# 确定自变量
y = f.iloc[:, 15]# 确定因变量
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1)# 根据q=√(n+m)+a确定隐含层，经测试隐含层在18的样本准确率最高
clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=12, random_state=1)
clf.fit(x_train, y_train)# 进行网络训练
rv = clf.score(x_train, y_train)# 获得神经网络的预测准确率（针对训练数据）
r = clf.predict(x_train)# 得出训练集的预测结果
R = clf.predict(x_test)# 得出测试集的预测结果

# 检测训练的结果
Rs = clf.score(x_test, y_test)

# 输出结果
print('模型预测的准确率为', rv)
print('预测的准确率为', Rs)
```
