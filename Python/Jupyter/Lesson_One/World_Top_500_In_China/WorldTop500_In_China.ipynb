```python
data['营业收入(百万美元)'] = data['营业收入(百万美元)'].str.replace(',','').astype('float')
```

```python
nameMap = {
    'China':'中国'
}
```

```python
x_data = data['字段2'].value_counts().index.tolist()
y_data = data['字段2'].value_counts().values.tolist()
```

```python
c = (
    Map()
    .add('', [list(z) for z in zip(x_data, y_data)], "china")
    .set_global_opts(title_opts=opts.TitleOpts(title="中国世界500强分布"),visualmap_opts=opts.VisualMapOpts( max_=data['字段2'].value_counts().max(),min_ = data['字段2'].value_counts().min(),
                                                        is_piecewise=False
                                                   ))
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
)
c.render_notebook()
```
