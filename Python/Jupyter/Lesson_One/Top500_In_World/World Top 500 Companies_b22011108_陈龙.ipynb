## 1.  ä¸–ç•Œ500å¼ºä¼ä¸šåˆ†å¸ƒåœ°å›¾

```python
x_data = data['å›½å®¶'].value_counts().index.tolist()
y_data = data['å›½å®¶'].value_counts().values.tolist()

# åœ°å›¾
map1 = Map()
map1.add('',[list(z) for z in zip(x_data, y_data)],
         maptype='world',is_map_symbol_show=False,
         name_map = nameMap
        ) 
map1.set_series_opts(label_opts=opts.LabelOpts(is_show=False))
map1.set_global_opts(title_opts=opts.TitleOpts(title='ä¸–ç•Œ500å¼ºä¼ä¸šåˆ†å¸ƒ'),
                     visualmap_opts=opts.VisualMapOpts( max_=data['å›½å®¶'].value_counts().max(),min_ = data['å›½å®¶'].value_counts().min(),
                                                       is_piecewise=False
                                                  ),
                    )
map1.render_notebook()
```

## 2.   2021å¹´å„å›½è¿›å…¥500å¼ºçš„ä¼ä¸šæ•°é‡æ’å

```python
result = data['å›½å®¶'].value_counts().sort_values()
x_data = result.index.tolist()
y_data = result.values.tolist()

bar = Bar(init_opts=opts.InitOpts(width = '900px',height = '900px'))
bar.add_xaxis(x_data)
bar.add_yaxis('',y_data)
bar.reversal_axis()
bar.set_series_opts(label_opts=opts.LabelOpts(position="right"))
bar.set_global_opts(title_opts=opts.TitleOpts(title="2021å¹´å„å›½è¿›å…¥500å¼ºä¼ä¸šæ•°é‡æ’å"))
bar.render_notebook()
```
